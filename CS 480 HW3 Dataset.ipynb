```python
%pylab inline
from collections import defaultdict
import requests, urllib
```

```python
url = 'https://data.boston.gov/api/3/action/datastore_search_sql'
```

```python
sql = '''
SELECT "DISTRICT",
       "OCCURRED_ON_DATE",
       "SHOOTING",
       "OFFENSE_DESCRIPTION"
FROM "b973d8cb-eeb2-4e7e-99da-c92938efc9c0"
'''
```

```python
general_url = url+"?sql="+urllib.parse.quote(sql)
```

```python
general_rows = requests.get( general_url ).json()
```

```python
general = general_rows['result']['records']
```

```python
# Only contains unique districts for the graph x-axis
districts = set()
for crime in general:
  # Avoids adding None and "External" districts
  if ((crime["DISTRICT"] != None) and (crime["DISTRICT"] != "External")):
    districts.add(crime["DISTRICT"])

# for district in districts:
#   print(district)
```

```python
# Chart 1 - Total crime count by district
districtToCount = {}

for crime in general:
  if ((crime["DISTRICT"] != None) and (crime["DISTRICT"] != "External")):
    districtToCount[crime["DISTRICT"]] = 1 + districtToCount.get(crime["DISTRICT"], 0)

districtToCount
```

```python
# Chart 2 - Crime types and its count for a specific district
# Maps District to a set of it's offense and frequency
# Ex: 'B3' maps to {'INVESTIGATE PERSON': 341}
districtToType = defaultdict(dict)

for crime in general:
  districtToType[crime["DISTRICT"]][crime["OFFENSE_DESCRIPTION"]] = 1 + districtToType[crime["DISTRICT"]].get(crime["OFFENSE_DESCRIPTION"], 0)

districtToType
```
